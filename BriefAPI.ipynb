## I.  Données Météorologiques actuelles
## Utilisez la clé API dans le script ci-dessous pour obtenir les données météorologiques actuelles de la ville de Paris.


```python
import requests
import pprintpp

#je definie une fonction get_current_weeather et je lui passe city et api_key en paramètres
def get_current_weather(city, api_key):
    base_url = "http://api.openweathermap.org/data/2.5/weather" # URL de l'API 
    # dictionnaire avec les paramètres clé:valeur
    params = {
        'q': city,
        'appid': api_key,
        'units': 'metric',  #  en degrés Celsius
        'lang': 'fr' # langue
    }
    # je stocke la réponse de l'API à la requête Get dans une variable nommée response
    response = requests.get(base_url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        weather = {
            'city': data['name'],
            'temperature': data['main']['temp'],
            'description': data['weather'][0]['description'],
            'humidity': data['main']['humidity'],
            'pressure': data['main']['pressure'],
            'wind_speed': data['wind']['speed']
        }
        return weather
    else:
        return None

# Ma clé API obtenue de OpenWeatherMap comme valeur de la variable api_key
api_key = 'c2c016ffe4ab841aa0fd81caf6b60d31'
city = 'Paris'

weather = get_current_weather(city, api_key)

if weather:
    print(f"Weather in {weather['city']}:")
    print(f"Temperature: {weather['temperature']}°C")
    print(f"Description: {weather['description']}")
    print(f"Humidity: {weather['humidity']}%")
    print(f"Pressure: {weather['pressure']} hPa")
    print(f"Wind Speed: {weather['wind_speed']} m/s")
else:
    print("Failed to get weather data")
```

## II. Prévisions météo à 5 jours


```python
import requests
from datetime import datetime

def get_forecast(city, api_key):
    base_url = "http://api.openweathermap.org/data/2.5/forecast"
    params = {
        'q': city,
        'appid': api_key,
        'units': 'metric',
        'lang': 'fr'
    }
    
    response = requests.get(base_url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        forecast = []
        for item in data['list']:
            weather_info = {
                'city': data['city']['name'],
                'temperature': item['main']['temp'],
                'temp_min': item['main']['temp_min'],
                'temp_max': item['main']['temp_max'],
                'pressure': item['main']['pressure'],
                'humidity': item['main']['humidity'],
                'weather': item['weather'][0]['description'],
                'wind_speed': item['wind']['speed'],
                'local_time': datetime.fromtimestamp(item['dt']).strftime('%Y-%m-%d %H:%M:%S')
            }
            forecast.append(weather_info)
        return forecast
    else:
        return None

#   ma clé API obtenue de OpenWeatherMap
api_key = 'c2c016ffe4ab841aa0fd81caf6b60d31'
cities = ['Paris', 'Lyon', 'Marseille']

for city in cities:
    forecast = get_forecast(city, api_key)

    if forecast:
        print(f"5-day/3-hour forecast for {city}:")
        for weather_info in forecast:
            print(f"Local Time: {weather_info['local_time']}")
            print(f"Temperature: {weather_info['temperature']}°C")
            print(f"Min Temperature: {weather_info['temp_min']}°C")
            print(f"Max Temperature: {weather_info['temp_max']}°C")
            print(f"Description: {weather_info['weather']}")
            print(f"Humidity: {weather_info['humidity']}%")
            print(f"Pressure: {weather_info['pressure']} hPa")
            print(f"Wind Speed: {weather_info['wind_speed']} m/s")
            print("\n")
    else:
        print(f"Failed to get forecast data for {city}")
```

## III. Données actuelles de pollution de l'air 

```python
import requests
from datetime import datetime
import pandas as pd

def get_air_pollution(city, api_key):
    # j'obtiens les coordonnées de la ville à partir de l'API de géocodage
    geocoding_url = "http://api.openweathermap.org/geo/1.0/direct"
    params = {
        'q': city,
        'limit': 1,
        'appid': api_key
    }
    response = requests.get(geocoding_url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        lat, lon = data[0]['lat'], data[0]['lon']
    else:
        return None

    # J'utilise ces coordonnées pour obtenir les données de pollution de l'air
    base_url = "http://api.openweathermap.org/data/2.5/air_pollution" # url de l'API
    params = {           # dictionnaire avec les paramètres clé:valeur
        'lat': lat,
        'lon': lon,
        'appid': api_key
    }
    
    response = requests.get(base_url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        pollution = {
            'city': city,
            'aqi': data['list'][0]['main']['aqi'],
            'co': data['list'][0]['components']['co'],
            'no': data['list'][0]['components']['no'],
            'no2': data['list'][0]['components']['no2'],
            'o3': data['list'][0]['components']['o3'],
            'so2': data['list'][0]['components']['so2'],
            'pm2_5': data['list'][0]['components']['pm2_5'],
            'pm10': data['list'][0]['components']['pm10'],
            'nh3': data['list'][0]['components']['nh3'],
            'local_time': datetime.fromtimestamp(data['list'][0]['dt']).strftime('%Y-%m-%d %H:%M:%S')
        }
        return pollution
    else:
        return None

# ma clé API obtenue de OpenWeatherMap comme valeur de la variable api_key
api_key = 'c2c016ffe4ab841aa0fd81caf6b60d31'
cities = ['Paris', 'Lyon', 'Marseille']

dataframes = []

for city in cities:
    pollution = get_air_pollution(city, api_key)

    if pollution:
        df = pd.DataFrame([pollution])
        dataframes.append(df)
    else:
        print(f"Failed to get air pollution data for {city}")

#  DataFrame pour la pollution de l'air
for df in dataframes:
    display(df)
```
